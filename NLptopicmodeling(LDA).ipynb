```python
data=pd.read_csv('articles.csv')
data.head()
```

```python
# Feature Extraction (Bag of Words)
abc= CountVectorizer(stop_words='english')
```

```python
x= abc.fit_transform(data)
x
```

```python
# Apply LDA (Latent Dirichlet Allocation)
topic=5
lda=LatentDirichletAllocation(n_components= topic, random_state=42)
```

```python
lda.fit(x)
```

```python
words = 5
feature= abc.get_feature_names_out()

def display(model,feature,words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"\nTopic{topic_idx+1}:")
        print(",".join([feature[i] for i in topic.argsort()[:-words - 1:-1]]))

display(lda,feature,words)
```
